```python
numerical_data = ["RT13V7", "RT13V8", "RT13V9","RT13V10","RT13V16"
,"RT13V19","RT13V22","RT13V25","RT13V34","RT13V37","RT13V40","RT13V43","RT13V46","RT13V49","RT13V52"]

truncatednumericaldata = ["RT13V16","RT13V19","RT13V22", "RT13V25","RT13V34","RT13V37","RT13V40","RT13V43","RT13V46","RT13V49","RT13V52"]
trunnumerical = df[numerical_data]
trunnum_cols=trunnumerical.columns
```

```python
numerical = df[numerical_data]
num_cols=numerical.columns
```

```python
categorical = df.drop(labels=numerical_data, axis =1)
cat_cols=categorical.columns
```

```python
#max value aka n/a values included
for col in num_cols:
    print(col)
    print('Skew :', round(df[col].skew(), 2))
    plt.figure(figsize = (15, 4))
    
    df[col].hist(grid=False)
    plt.ylabel('count')
    plt.subplot(1, 2, 2)
    sns.boxplot(x=df[col])
    plt.show()
```

```python
#Manually truncate data:
trundf = df[df["RT13V7"] < 95]
trundf["RT13V8"][trundf["RT13V8"] == 95] = 0
trundf = trundf[trundf["RT13V8"] < 98]
for col in trunnum_cols:
    maxcol = max(trundf[col])
    trundf = trundf[trundf[col] < maxcol]


```

```python
for col in num_cols:
    print(col)
    print('Skew :', round(trundf[col].skew(), 2))
    plt.figure(figsize = (15, 4))
    plt.subplot(1, 2, 1)
    trundf[col].hist(grid=False)
    plt.ylabel('count')
    plt.subplot(1, 2, 2)
    sns.boxplot(x=trundf[col])
    plt.show() 
```

```python
cat_colst = cat_cols.drop(['ER53002','RT13V2'])
```
